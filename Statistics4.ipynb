```python
Q1: Probability Mass Function (PMF) and Probability Density Function (PDF)

Probability Mass Function (PMF):
The Probability Mass Function (PMF) is used for discrete random variables. It gives the probability of a specific outcome occurring. The PMF maps each possible value of the random variable to its associated probability. The sum of all the probabilities in the PMF is equal to 1.

Example of PMF:
Consider a fair six-sided die. The PMF for rolling the die is as follows:

P(X = 1) = 1/6
P(X = 2) = 1/6
P(X = 3) = 1/6
P(X = 4) = 1/6
P(X = 5) = 1/6
P(X = 6) = 1/6
Probability Density Function (PDF):
The Probability Density Function (PDF) is used for continuous random variables. Unlike the PMF, the PDF does not give the probability of a specific value but provides the relative likelihood of the random variable taking a range of values. The area under the PDF curve within a certain range represents the probability of the random variable falling within that range.

Example of PDF:
The PDF of a standard normal distribution (mean = 0, standard deviation = 1) is given by the bell-shaped curve known as the standard normal curve.

Q2: Cumulative Density Function (CDF)

The Cumulative Density Function (CDF) is a function associated with a probability distribution that provides the cumulative probability that a random variable takes on a value less than or equal to a given value. The CDF is defined for both discrete and continuous random variables.

Example of CDF:
For a fair six-sided die, the CDF is as follows:

F(X ≤ 1) = 1/6
F(X ≤ 2) = 2/6 = 1/3
F(X ≤ 3) = 3/6 = 1/2
F(X ≤ 4) = 4/6 = 2/3
F(X ≤ 5) = 5/6
F(X ≤ 6) = 1
The CDF provides information about the likelihood of a random variable being less than or equal to a given value. It is used in statistical analysis to compute probabilities and make probability-based decisions.

Q3: Examples of situations where the normal distribution might be used as a model:

In natural phenomena, such as the distribution of human heights, weights, and IQ scores.
In manufacturing processes, such as the distribution of product lengths or weights.
In financial markets, to model stock returns or asset prices.
In quality control, to monitor production variations.
The parameters of the normal distribution (mean and standard deviation) relate to the shape of the distribution:

The mean (μ) determines the central location or peak of the distribution.
The standard deviation (σ) controls the spread or width of the distribution. A larger standard deviation results in a wider and flatter distribution, while a smaller standard deviation leads to a narrower and taller distribution.
Q4: Importance of Normal Distribution and Real-life examples:
The normal distribution is of great importance in statistics and probability theory for several reasons:

It is mathematically tractable and analytically convenient, making it easier to perform calculations and inferential statistics.
It serves as an excellent approximation for many real-world phenomena, thanks to the Central Limit Theorem.
Many statistical methods and hypothesis tests are built on the assumption of normality.
Examples of real-life applications of the normal distribution:

IQ scores: IQ scores are often assumed to follow a normal distribution, with the mean set at 100 and a standard deviation of 15.
Heights and weights of individuals in a population: In large populations, the distribution of heights and weights often approximates a normal distribution.
Test scores: The scores on standardized tests, such as SAT or GRE, are often assumed to follow a normal distribution.
Q5: Bernoulli Distribution and Binomial Distribution

Bernoulli Distribution:

The Bernoulli distribution models a random experiment with two possible outcomes, usually labeled as success (S) with probability p and failure (F) with probability (1-p).
The random variable X represents whether the event occurs (1 for success, 0 for failure).
It is used to model single trials or events with a binary outcome.
Example of Bernoulli Distribution:
A coin toss experiment, where the outcome is either heads (success) or tails (failure) with a certain probability.

Difference between Bernoulli Distribution and Binomial Distribution:

Bernoulli Distribution deals with a single trial, while Binomial Distribution deals with multiple independent and identical trials.
Bernoulli Distribution has one parameter (probability of success, p), whereas Binomial Distribution has two parameters (number of trials, n, and probability of success, p).
In a Binomial Distribution, the random variable X represents the number of successes in n trials, whereas in Bernoulli, it represents the outcome of a single trial (0 or 1).
Q6: Probability that a randomly selected observation will be greater than 60 for a normal distribution:

Let's assume the dataset is normally distributed with a mean (μ) of 50 and a standard deviation (σ) of 10. We need to find the probability that a randomly selected observation (X) will be greater than 60.

Using the Z-score formula:
Z = (X - μ) / σ

Z = (60 - 50) / 10
Z = 1

Now, using the standard normal distribution table or calculator, we find the probability associated with a Z-score of 1, which is approximately 0.8413.

So, the probability that a randomly selected observation will be greater than 60 is approximately 0.8413 or 84.13%.

Q7: Uniform Distribution

Uniform Distribution:

The uniform distribution is a probability distribution where all values in a given range have equal probability.
It is characterized by a constant probability density over the entire range and a rectangular-shaped probability density function.
Example of Uniform Distribution:
Consider rolling a fair six-sided die. Each face has an equal probability of 1/6 of showing up.

Q8: Z-score and its Importance

Z-score:

The Z-score (also known as the standard score) is a standardized value that represents the number of standard deviations a data point is from the mean.
It helps in comparing individual data points from different normal distributions on a common scale.
Importance of Z-score:

It allows us to determine how unusual or typical a data point is in comparison to the rest of the distribution.
It helps identify outliers and extreme values.
It is used in hypothesis testing and statistical analysis to determine the probability of observing a data point in a standard normal distribution.
Q9: Central Limit Theorem (CLT)

Central Limit Theorem:

The Central Limit Theorem states that the sampling distribution of the sample means from a large sample drawn from any population will be approximately normally distributed, regardless of the shape of the original population distribution.
As the sample size increases, the sampling distribution of the sample mean approaches a normal distribution, even if the population distribution is not normal.
Significance of the Central Limit Theorem:

It forms the basis for many statistical methods and inference techniques.
It allows statisticians to make assumptions about the sampling distribution of the sample mean, making hypothesis testing and confidence interval estimation more feasible.
Q10: Assumptions of the Central Limit Theorem

The Central Limit Theorem assumes the following:

The sample is large enough (typically n ≥ 30).
The observations in the sample are independent of each other.
The sample is selected randomly from the population of interest.
The population from which the sample is drawn may or may not be normally distributed. However, for small samples or highly skewed populations, the CLT may require a larger sample size to be effective
```
